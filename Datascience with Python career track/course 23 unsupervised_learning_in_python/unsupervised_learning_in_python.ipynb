### Evaluating a clustering
- check against provided clusters/labels (but maybe not available)
- make confusion matric with each cluster label and given label
- If labels not available: itertia
- inertia measures how far are the samples from the centroids (model.inertia_)
- tradeoff: a good clustering has low inertia(tight clusters) but not too many clusters
- rule of thumb is to choose the elbow in inertia curve(inertia vs k)

````
import pandas as pd
df = pd.read_csv('seeds.csv',header=None)
df.head()
````

````
samples = df[[i for i in range(7)]].values
len(df), samples.shape
````

````
ks = range(1, 6)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(samples)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
````

### Intrinsic dimension:
- number of features needed to approximate the dataset
- this is the essential idea behind dimension reduction
- to find out what is the most compact representation of the dataset?
- intrinsic features: the dimentions that have high variance are counted (rest are almost same, i.e low variance so we can approximate them)
- pca.explained_variance_

The first principal component of the data is the direction in which the data varies the most.

````
# Make a scatter plot of the untransformed points
plt.scatter(grains.loc[:,3], grains.loc[:,4])

# Create a PCA instance: model
model = PCA()

# Fit model to points
model.fit(grains)

# Get the mean of the grain samples: mean
mean = model.mean_

# Get the first principal component: first_pc
first_pc = model.components_[0,:]

# Plot first_pc as an arrow, starting at mean
plt.arrow(mean[0], mean[1], first_pc[0], first_pc[1], color='red', width=0.01)

# Keep axes on same scale
plt.axis('equal')
plt.show()
````

# Discovering interpretable features

## NMF
- non negative matrix factorization
- models are interpretable(unlike SVD)
- dimentionality reduction technique
- all sample features should be >=0
- nmf decomposes documents as combination of common theme, and images as combination of common patterns
- **must specify number of components**
- works with both numoy array and csr matrix
- dimention of components = dimention of samples
- inp = (doc by words)
- nmf.components_ = (topic by words)
- nmf_features(output) = (doc by topic)
- features*components = reconstructed samples

### NMF on wikipedia

````
import pandas as pd
from scipy.sparse import csr_matrix

df = pd.read_csv('wikipedia-vectors.csv', index_col=0)
articles = csr_matrix(df.transpose())
titles = list(df.columns)
titles[:10]
````
