# Pandas
need to store tabular data, but numpy arrays can have only a single type of data, pandas helps, a package on top of numpy :)

can create a dataframe from key-value pairs

```
import pandas as pd
```

#### dict to df

```
data = {
    'country' : ['spain', 'france', 'germany', 'norway'],
    'capital' : ['madrid', 'paris', 'berlin', 'oslo']
}
```

```
df = pd.DataFrame(data)
df
```

```
df.index=['SP','FR','GE','NR']
df
```

```
# Pre-defined lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

import pandas as pd

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {
    'country':names,
    'drives_right':dr,
    'cars_per_cap':cpc
}

# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict)

# Print cars
print(cars)
```

# looping

### looping in dicts

```
data = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }
```

```
for a,b in data.items():
    print(a,b)
```

### looping in n-dim numoy arrays

```
x = np.random.rand(3)
y = np.random.rand(3)
```

```
z=np.array([x,y])
z.shape
```

```
for arr in z: #prints entire 1d arrays
    print(arr)
```

```
for arr in np.nditer(z): #loops element wise
    print(arr)
```

```
cars
```

```
## looping through each row
df = cars.head(1)
for label,data in df.iterrows():
    print("label = " + str(label))
    print(type(data))
    print(data)
```
